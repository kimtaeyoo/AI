```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

nbastat = pd.read_csv('nbastat2022c.csv')
m = len(nbastat)

x = nbastat[['FGA','3PA','FTA']]
y = nbastat['PTS']

x = x.dropna()
y = y.dropna()

x = (np.array(x)).reshape(m, 3)
y = (np.array(y)).reshape(m, 1)

learning_rate = 0.0001
n_iter = 2000

theta = np.zeros((4,1))
gradients = np.zeros((4,1))
x0 = np.ones((m, 1))
xb = np.c_[x0,x]

losses = []
for i in range(n_iter):
    loss = ((1/(2.0*m))*(xb.dot(theta)-y)**2).sum()
    losses.append(loss)
    gradients = (1/m)*xb.T.dot(xb.dot(theta)-y)
    theta = theta - learning_rate * gradients

plt.plot(losses)
```
